```python
def replace_config_value(file_path, search_string, replacement):
    with open(file_path, 'r') as file:
        file_content = file.read()

    updated_content = file_content.replace(search_string, replacement)

    with open(file_path, 'w') as file:
        file.write(updated_content)

replace_config_value('nginx.conf', 'server_01', server_01)
```
